### Get Info from tensor 

There will be times when you'll want to get different pieces of information from your tensors, in particuluar, you should know the following tensor vocabulary:

Shape: The length (number of elements) of each of the dimensions of a tensor.
Rank: The number of tensor dimensions. A scalar has rank 0, a vector has rank 1, a matrix is rank 2, a tensor has rank n.
Axis or Dimension: A particular dimension of a tensor.
Size: The total number of items in the tensor.

#### find the positional max and min(argmax and argmin)
##### where to use --->

The Primary Use Case for argmin
While argmax is used for finding the most confident prediction, argmin is often used when dealing with Distance or Loss.

K-Nearest Neighbors / Clustering: If you are building a recommendation engine and want to find the user whose profile is closest to yours, you calculate the mathematical distance between you and everyone else. You use argmin to find the index of the user with the minimum distance.

Vocabulary/Embedding Lookups: In Natural Language Processing (NLP), if the model outputs a word vector, you compare it to a dictionary of 10,000 known words. You use argmin to find the index of the word that has the smallest difference from the model's output.

```python
tf.random.set_seed(42)
F = tf.random.uniform(shape=[50])
F
```

```python
idx = tf.argmax(F)
idx
```

```python
F[idx]
```

```python
tf.reduce_max(F)
```

```python
F[idx] == tf.reduce_max(F)
```

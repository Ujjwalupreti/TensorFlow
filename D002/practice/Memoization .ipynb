#### Memoization

a powerful optimization technique that can drastically improve the performance of certain algorithms. 

Memoization helps by storing the results of expensive function calls and reusing them when the same inputs occur again. 

This avoids redundant calculations, making your code more efficient

#### How MLPs use Memoization (The Forward-Backward Cache)


If you do not use memoization, training a Neural Network would take years instead of hours.

When an MLP does a Forward Pass, it calculates the math for Layer 1, then Layer 2, then Layer 3.
When it does the Backward Pass to calculate the gradients, the calculus Chain Rule explicitly requires the numbers that were generated during the Forward Pass!

If you didn't memoize (cache) those numbers, the network would have to completely recalculate the Forward Pass every single time it took a step backward.

```python
# Small function from Backpropagation.py 

def __mul__(self, other):
        # 1. Calculate the math once
        out_data = self.data * other.data 
        
        # 2. MEMOIZE IT! Save it in a new Value object
        out = Value(out_data, (self, other), '*')

        def _backward():
            # 3. USE THE CACHE! 
            # When backprop happens later, we don't recalculate self * other.
            # We just grab the memoized 'other.data' and 'self.data' from memory.
            self.grad += other.data * out.grad
            other.grad += self.data * out.grad
```

Look at your Value class. When you do a forward calculation like x * y, you create a new Value object that stores self.data.

self.data is your memoization cache!
